## 🎯 Bu Notebook'ta Ne Öğreneceğiz?

İlk notebook'ta tek bir nöron gördük. Şimdi sıra **çok katmanlı sinir ağında**.

Sonunda şunları bileceksiniz:

- 🧱 **Katman (Layer)** kavramı — nöronları nasıl yan yana, art arda diziyoruz?
- 🔢 **Sınıflandırma** problemleri — sadece sayı değil, **kategori** tahmin etmek
- 🧪 **Train/Test ayrımı** — modelin gerçekten öğrenip öğrenmediğini nasıl ölçeriz?
- 📊 **Doğruluk (Accuracy)** ne demek?
- ⚙️ **Keras** ile sinir ağını **3 satırda** nasıl kurarız?


## 🤔 Önce Bir Senaryo

Bir botanikçi düşünün. Iris (süsen) çiçeğinin **3 türü** var:

- 🌸 **Setosa**
- 🌺 **Versicolor**
- 🌻 **Virginica**

Her çiçeğin 4 ölçüsü var:
1. Sepal (taç yaprak) uzunluğu
2. Sepal genişliği
3. Petal (çanak yaprak) uzunluğu
4. Petal genişliği

**Soru:** Bu 4 ölçüye bakarak çiçeğin türünü bilebilir miyiz?

Bu **klasik** bir makine öğrenmesi problemidir — 1936'dan beri üzerinde çalışılıyor!

---

### 🧱 Tek Nöron Yetmez Mi?

Pizza örneğinde bir sayı tahmin ediyorduk. Burada **3 sınıftan birini** seçmemiz lazım. Ayrıca **4 girdimiz** var.

Cevap: **Birden fazla nöron + birden fazla katman** kullanacağız.

> 🎓 **Analoji:** Bir basketbol takımı düşünün. Tek bir oyuncu maçı kazanamaz. Pas atan, top toplayan, sayı atan farklı pozisyonlar olur. Sinir ağındaki katmanlar da böyle: **her katman farklı bir iş yapar**, sondaki sonucu birlikte üretirler.


## 📦 Kurulum

Bu notebook'ta TensorFlow / Keras kullanacağız.

> 💬 Google Colab'da bunlar **zaten yüklü**. Aşağıdaki hücreyi çalıştırın yeter.

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# scikit-learn — meşhur makine öğrenmesi kütüphanesi (Iris veri seti burada hazır geliyor)
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

# Keras — derin öğrenme kütüphanesi
from tensorflow import keras
from tensorflow.keras import layers

# Tekrarlanabilir sonuçlar için
np.random.seed(42)
keras.utils.set_random_seed(42)

print("✅ Kütüphaneler hazır!")
print(f"TensorFlow / Keras sürümü: {keras.__version__}")
````

## 📊 Veriyi Tanıyalım

Iris veri seti **scikit-learn** içinde hazır geliyor. Tek satırla yüklüyoruz:

````python
# Veri setini yükle
iris = load_iris()

X = iris.data        # 4 ölçü (özellikler)
y = iris.target      # 3 sınıf (0, 1, 2)
ozellik_isimleri = iris.feature_names
sinif_isimleri = iris.target_names

print(f"📦 Veri boyutu: {X.shape}    →  {X.shape[0]} çiçek, her birinde {X.shape[1]} ölçü")
print(f"🏷️  Etiket boyutu: {y.shape}")
print(f"🌸 Sınıflar: {sinif_isimleri}")
print()
print(f"İlk 5 örnek:")
print("Sepal Uz | Sepal Gen | Petal Uz | Petal Gen | Sınıf")
print("-" * 55)
for i in range(5):
    print(f"  {X[i,0]:>6.1f} |   {X[i,1]:>6.1f}  |  {X[i,2]:>6.1f}  |   {X[i,3]:>6.1f}  |  {sinif_isimleri[y[i]]}")
````

### Sınıfların Dengeli Olduğunu Kontrol Edelim

> 💡 **Önemli:** Bir sınıftan çok daha az örnek varsa model o sınıfı iyi öğrenemez!

````python
sayilar = np.bincount(y)

fig, ax = plt.subplots(figsize=(8, 4))
renkler = ['#E91E63', '#9C27B0', '#FFC107']
ax.bar(sinif_isimleri, sayilar, color=renkler, edgecolor='white', linewidth=2)
for i, s in enumerate(sayilar):
    ax.text(i, s + 1, str(s), ha='center', fontweight='bold', fontsize=12)
ax.set_ylabel('Örnek Sayısı')
ax.set_title('🌸 Iris Sınıf Dağılımı', fontweight='bold')
plt.show()

print("✅ Her sınıftan eşit (50'şer) örnek var. Veri dengeli!")
````

### Görsel Keşif — Hangi Ölçü Hangi Sınıfı Ayırıyor?

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Sepal uzunluk vs sepal genişlik
for i, isim in enumerate(sinif_isimleri):
    mask = y == i
    axes[0].scatter(X[mask, 0], X[mask, 1], label=isim, color=renkler[i], s=60, edgecolor='white')
axes[0].set_xlabel('Sepal Uzunluğu (cm)')
axes[0].set_ylabel('Sepal Genişliği (cm)')
axes[0].set_title('Sepal Ölçüleri')
axes[0].legend()

# Petal uzunluk vs petal genişlik
for i, isim in enumerate(sinif_isimleri):
    mask = y == i
    axes[1].scatter(X[mask, 2], X[mask, 3], label=isim, color=renkler[i], s=60, edgecolor='white')
axes[1].set_xlabel('Petal Uzunluğu (cm)')
axes[1].set_ylabel('Petal Genişliği (cm)')
axes[1].set_title('Petal Ölçüleri')
axes[1].legend()

plt.tight_layout()
plt.show()

print("👀 Petal ölçüleri sınıfları daha net ayırıyor — model bunları daha çok kullanacak.")
````

## 🧪 Veriyi Eğitim ve Test Olarak Ayıralım

**Neden ayırıyoruz?** Çünkü modelin **gerçekten öğrenip öğrenmediğini** anlamamız lazım. Eğer modele eğittiğimiz örnekleri test edersek, sadece **ezberlemiş** olabilir!

> 🎓 **Analoji:** Öğrencilere sınavda **çıkmış soruları** verirseniz, bilen ile ezberleyen arasındaki farkı göremezsiniz. Yeni sorular sormak gerekir.

Veriyi **%80 eğitim / %20 test** olarak ayırıyoruz:

````python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,        # %20'si test
    random_state=42,      # Aynı sonucu görmek için
    stratify=y            # Her sınıftan oransal alır
)

print(f"📚 Eğitim: {X_train.shape[0]} örnek")
print(f"📝 Test  : {X_test.shape[0]} örnek")
````

## 🧠 Sinir Ağını Kuralım

Modelimiz **3 katmanlı** olacak:

```
[4 girdi] → [8 nöronlu gizli katman] → [8 nöronlu gizli katman] → [3 nöronlu çıktı katmanı]
```

- **Gizli katmanlar** ReLU aktivasyonu kullanır (negatifleri sıfırlar, pozitifleri olduğu gibi geçirir)
- **Çıktı katmanı** Softmax kullanır — 3 sınıfın olasılıklarını verir, toplamları 1 olur

> 🎓 **Aktivasyon fonksiyonu nedir?** Nöronun ürettiği sayıyı son şekle sokan filtre. ReLU, "negatifsen 0 ol" demek. Softmax ise "olasılıklara çevir" demek.


````python
# Keras Sequential API: katmanları sırayla diziyoruz
model = keras.Sequential([
    layers.Input(shape=(4,)),                       # 4 girdi (4 ölçü)
    layers.Dense(8, activation='relu', name='gizli_1'),
    layers.Dense(8, activation='relu', name='gizli_2'),
    layers.Dense(3, activation='softmax', name='cikti')  # 3 sınıf
])

# Modelin nasıl göründüğünü yazdır
model.summary()
````

### Eğitim Ayarlarını Belirleyelim

Modele şunları söylüyoruz:

- **Optimizer (Adam):** Ağırlıkları nasıl güncelleyeceği. Adam yaygın ve iyi bir seçim.
- **Loss (sparse_categorical_crossentropy):** Sınıflandırmada standart hata ölçümü.
- **Metrics (accuracy):** Doğruluk yüzdesi takip edilsin.

````python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("✅ Model derlendi (compile edildi). Artık eğitime hazır!")
````

## 🚀 Eğitim Zamanı!

`model.fit()` ile modeli eğitiyoruz. **100 epoch** boyunca tüm eğitim verisini gözden geçirecek.

````python
gecmis = model.fit(
    X_train, y_train,
    epochs=100,
    validation_split=0.2,    # Eğitim verisinin %20'si doğrulama için ayrılsın
    verbose=0                # Her epoch'u yazdırma (kalabalık olmasın)
)

print(f"🎉 Eğitim tamamlandı!")
print(f"   Son eğitim doğruluğu : {gecmis.history['accuracy'][-1] * 100:.1f}%")
print(f"   Son doğrulama doğruluğu: {gecmis.history['val_accuracy'][-1] * 100:.1f}%")
````

### 📈 Eğitim Süreci Grafiği

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Doğruluk
axes[0].plot(gecmis.history['accuracy'], label='Eğitim', color='#1E88E5', linewidth=2)
axes[0].plot(gecmis.history['val_accuracy'], label='Doğrulama', color='#E53935', linewidth=2)
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Doğruluk')
axes[0].set_title('📈 Doğruluk Artıyor', fontweight='bold')
axes[0].legend()

# Hata
axes[1].plot(gecmis.history['loss'], label='Eğitim', color='#1E88E5', linewidth=2)
axes[1].plot(gecmis.history['val_loss'], label='Doğrulama', color='#E53935', linewidth=2)
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Hata (Loss)')
axes[1].set_title('📉 Hata Azalıyor', fontweight='bold')
axes[1].legend()

plt.tight_layout()
plt.show()
````

## 🔍 Test Verisinde Modeli Sınayalım

Modelin **hiç görmediği** test verisi üzerinde performansı:

````python
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
print(f"🎯 Test doğruluğu: {test_acc * 100:.1f}%")
print(f"   Yani {len(X_test)} test örneğinden {int(test_acc * len(X_test))} tanesi doğru tahmin edildi.")
````

### Confusion Matrix — Hangi Sınıfı Karıştırıyor?

Bu matris, modelin **hangi sınıfı hangi sınıfla karıştırdığını** gösterir.

````python
# Test setindeki tahminler
tahmin_olasiliklari = model.predict(X_test, verbose=0)
y_pred = np.argmax(tahmin_olasiliklari, axis=1)

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(7, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=sinif_isimleri, yticklabels=sinif_isimleri,
            cbar=False, annot_kws={'size': 14, 'weight': 'bold'})
plt.xlabel('Tahmin')
plt.ylabel('Gerçek')
plt.title('🧩 Confusion Matrix — Modelin Karıştırdığı Sınıflar', fontweight='bold')
plt.show()

print("\n📋 Detaylı rapor:")
print(classification_report(y_test, y_pred, target_names=sinif_isimleri))
````

## 🎮 Kendin Dene

Aşağıda kendi ölçülerinizi girip çiçek türü tahmini alabilirsiniz:

````python
# 👇 Kendi ölçülerinizi yazın (cm cinsinden)
sepal_uzunluk  = 5.5
sepal_genislik = 3.0
petal_uzunluk  = 4.0
petal_genislik = 1.3

# Modele veriyi sok
girdi = np.array([[sepal_uzunluk, sepal_genislik, petal_uzunluk, petal_genislik]])
olasiliklar = model.predict(girdi, verbose=0)[0]
tahmin_sinif = np.argmax(olasiliklar)

print(f"🌸 Tahmini Tür: {sinif_isimleri[tahmin_sinif].upper()}")
print()
print("Detaylı olasılıklar:")
for isim, olasilik in zip(sinif_isimleri, olasiliklar):
    bar = '█' * int(olasilik * 30)
    print(f"  {isim:<12} {bar} {olasilik*100:.1f}%")
````

## 📝 Özet — Bu Notebook'ta Ne Öğrendik?

| **Softmax** | Olasılık Çıktısı | Her sınıfa olasılık atayan fonksiyon |

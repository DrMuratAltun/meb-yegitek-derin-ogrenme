Her adım, **bir öncekinin çıktısını** girdi olarak alır. Hiçbir adım atlanmaz, sıra bellidir. **Yön: ileri.**

Sinir ağı tam olarak böyle çalışır. Veri, **katman katman**, her katmanda matematik işlemi görerek **sondaki tahmine** ulaşır. Buna **İleri Yayılım** diyoruz.

> 🎓 **Anahtar:** Bir katmanın çıktısı, bir sonrakinin girdisidir. Adımlar **sırayla** ve **ileri** yöndedir.


## 📦 Kurulum

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
plt.rcParams['figure.figsize'] = (10, 5)
print("✅ Hazır!")
````

## 1️⃣ Tek Nöron → Tek Katmandaki Birden Çok Nöron

Notebook 01'de bir nöron şöyleydi:

$$y = w \cdot x + b$$

Şimdi **3 nöron**lu tek bir katman düşünelim. Her birinin **kendi w'si ve b'si var**:

$$y_1 = w_1 \cdot x + b_1$$
$$y_2 = w_2 \cdot x + b_2$$
$$y_3 = w_3 \cdot x + b_3$$

Üçünü ayrı ayrı yazmak yerine **tek satıra** sığdırabilir miyiz? Evet — **matris ile**:

$$\begin{bmatrix} y_1 \\ y_2 \\ y_3 \end{bmatrix} = \begin{bmatrix} w_1 \\ w_2 \\ w_3 \end{bmatrix} \cdot x + \begin{bmatrix} b_1 \\ b_2 \\ b_3 \end{bmatrix}$$

> 🎓 **Matris çarpımı = paralel hesap.** 100 nöronu tek tek hesaplamak yerine bir matris işlemiyle yaparız. Bu yüzden GPU'lar sinir ağlarında bu kadar hızlıdır.


````python
# Örnek: 1 girdi, 3 nöronlu tek katman
x = 2.5

# Her nöronun w ve b'si
W = np.array([[0.5], [-0.3], [0.8]])    # Şekil: (3, 1) — 3 nöron, her biri 1 girdi alır
b = np.array([[0.1], [0.0], [-0.2]])    # Şekil: (3, 1)

# Tek satır hesap
y = W * x + b

print("Girdi x:", x)
print("\n3 nöronun ağırlıkları (W):"); print(W)
print("\nBias'lar (b):"); print(b)
print("\n3 nöronun çıktıları (y):"); print(y)
````

## 2️⃣ Birden Fazla Girdi: Gerçek Matris Çarpımı

Çoğu zaman **birden fazla girdimiz** olur. Iris çiçeğinde 4 ölçü vardı, fotoğrafta 784 piksel olur.

Diyelim ki:
- **2 girdi** ($x_1, x_2$) — örneğin pizza çapı ve kalınlığı
- **3 nöronlu katman**

O zaman **W matrisi (3 satır × 2 sütun)** olur:

$$W = \begin{bmatrix} w_{11} & w_{12} \\ w_{21} & w_{22} \\ w_{31} & w_{32} \end{bmatrix}, \quad x = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}, \quad b = \begin{bmatrix} b_1 \\ b_2 \\ b_3 \end{bmatrix}$$

İleri yayılım:

$$z = W \cdot x + b$$

Sonra aktivasyondan geçer:

$$a = \text{ReLU}(z)$$

````python
# 2 girdi, 3 nöronlu katman
def relu(x):
    return np.maximum(0, x)

x = np.array([[1.5], [2.0]])     # Girdi: (2, 1)

W = np.array([
    [ 0.4, -0.2],     # 1. nöron
    [-0.1,  0.5],     # 2. nöron
    [ 0.3,  0.6],     # 3. nöron
])                                # (3, 2)

b = np.array([[0.1], [0.0], [-0.1]])   # (3, 1)

# İleri yayılım
z = W @ x + b      # @ matris çarpımı demek
a = relu(z)

print("Girdi x (şekil {}):".format(x.shape))
print(x)
print("\nAğırlık matrisi W (şekil {}):".format(W.shape))
print(W)
print("\nLineer çıkış z = W·x + b (şekil {}):".format(z.shape))
print(z)
print("\nAktivasyon sonrası a = ReLU(z):")
print(a)
````

## 3️⃣ Çok Katmanlı Ağ — Adım Adım

Şimdi **2 katmanlı** bir ağ kuralım. Bu, "çok katmanlı algılayıcı (MLP)" diye geçer.

```
Girdi (2)  →  Gizli Katman (4 nöron, ReLU)  →  Çıkış Katmanı (1 nöron, Sigmoid)
```

Yapacaklarımız:

| Adım | Hesap | Şekil |
|---|---|---|
| 1 | $z_1 = W_1 \cdot x + b_1$ | (4, 1) |
| 2 | $a_1 = \text{ReLU}(z_1)$ | (4, 1) |
| 3 | $z_2 = W_2 \cdot a_1 + b_2$ | (1, 1) |
| 4 | $a_2 = \sigma(z_2)$ | (1, 1) |

Çıktı $a_2$, **0 ile 1 arasında** olacak (ikili sınıflandırma için olasılık).

````python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Bir örnek girdi
x = np.array([[1.0], [0.5]])

# Katman 1: 2 girdi → 4 nöron
W1 = np.random.randn(4, 2) * 0.5
b1 = np.zeros((4, 1))

# Katman 2: 4 girdi → 1 nöron
W2 = np.random.randn(1, 4) * 0.5
b2 = np.zeros((1, 1))

# Adım 1: Lineer
z1 = W1 @ x + b1
print(f"Adım 1: z1 = W1·x + b1   →  şekil {z1.shape}")
print(z1.round(3))

# Adım 2: ReLU
a1 = relu(z1)
print(f"\nAdım 2: a1 = ReLU(z1)   →  şekil {a1.shape}")
print(a1.round(3))

# Adım 3: Lineer
z2 = W2 @ a1 + b2
print(f"\nAdım 3: z2 = W2·a1 + b2  →  şekil {z2.shape}")
print(z2.round(3))

# Adım 4: Sigmoid
a2 = sigmoid(z2)
print(f"\nAdım 4: a2 = σ(z2)       →  şekil {a2.shape}")
print(a2.round(3))

print(f"\n🎯 Sonuç: Ağın tahmin ettiği olasılık = {a2[0,0]:.3f}")
````

### Veri Akışını Görselleştirelim

````python
# Yatay bar grafiklerle her katmandaki aktivasyonları görelim
fig, axes = plt.subplots(1, 4, figsize=(16, 3.5))

axes[0].barh(['x1', 'x2'], x.flatten(), color='#1E88E5')
axes[0].set_title('Girdi (x)', fontweight='bold')

axes[1].barh([f'z1[{i}]' for i in range(4)], z1.flatten(),
             color=['#43A047' if v > 0 else '#E53935' for v in z1.flatten()])
axes[1].set_title('z1 = W1·x + b1', fontweight='bold')
axes[1].axvline(x=0, color='black', alpha=0.3)

axes[2].barh([f'a1[{i}]' for i in range(4)], a1.flatten(), color='#43A047')
axes[2].set_title('a1 = ReLU(z1)', fontweight='bold')

axes[3].barh(['Tahmin'], a2.flatten(), color='#FF6F00')
axes[3].set_xlim([0, 1])
axes[3].set_title(f'a2 = σ(z2) = {a2[0,0]:.3f}', fontweight='bold')

plt.tight_layout()
plt.show()

print("👀 Her katmanda veri 'şekil değiştirerek' ilerliyor. Bu İLERİ YAYILIM!")
````

## 4️⃣ Mini Bir Veri Setiyle Toplu İleri Yayılım

Tek örnek değil, **birden fazla örneği aynı anda** işleyebiliriz. Buna **batch işleme** denir.

X'in şekli (2, N) olur — N örnek, her birinin 2 özelliği. Ağdan toplu geçer.

````python
# 4 örnekten oluşan bir batch
X = np.array([
    [0.0, 0.0, 1.0, 1.0],   # x1 değerleri
    [0.0, 1.0, 0.0, 1.0],   # x2 değerleri
])
print("Batch X (şekil {}):".format(X.shape))
print(X)

# Tüm batch'i tek seferde geçir
Z1 = W1 @ X + b1
A1 = relu(Z1)
Z2 = W2 @ A1 + b2
A2 = sigmoid(Z2)

print("\nAğın 4 örnek için tahminleri:")
for i in range(4):
    print(f"  Girdi ({X[0,i]}, {X[1,i]}) → Olasılık {A2[0,i]:.3f}")
````

## 5️⃣ XOR Örneği — Aktivasyonun Gerçek Gücü

Önceki notebook'ta gösterdik: **XOR doğrusal olarak ayrılamaz**. Aktivasyonsuz ağ çözemez.

Şimdi **manuel olarak** doğru ağırlıkları **biz seçelim** (öğrenmeyi 5. notebook'ta göreceğiz). Amaç: ileri yayılımın bunu çözebileceğini görmek.

````python
# XOR için elle ayarlanmış ağırlıklar
W1_xor = np.array([
    [ 1,  1],   # OR-benzeri
    [-1, -1],   # NOR-benzeri (ters OR)
])
b1_xor = np.array([[-0.5], [1.5]])

W2_xor = np.array([[1, 1]])
b2_xor = np.array([[-1.5]])

# 4 XOR örneği
X_xor = np.array([
    [0, 0, 1, 1],
    [0, 1, 0, 1],
])
y_gercek = [0, 1, 1, 0]    # XOR gerçek değerleri

# İleri yayılım
Z1 = W1_xor @ X_xor + b1_xor
A1 = relu(Z1)
Z2 = W2_xor @ A1 + b2_xor
A2 = (Z2 > 0).astype(int)   # Step fonksiyonu (basitlik için)

print("Girdi (x1, x2) | Beklenen | Ağ Tahmini")
print("-" * 40)
for i in range(4):
    print(f"   ({X_xor[0,i]}, {X_xor[1,i]})       |    {y_gercek[i]}     |    {A2[0,i]}")

print("\n✅ XOR çözüldü! 1 katman lineer çözemiyordu, 2 katman + ReLU yapabiliyor.")
````

## 🆚 Aynı Şeyi Keras ile Yapsak?

Yukarıdaki tüm hesabı NumPy ile elle yazdık. Keras'ta görelim:

````python
from tensorflow import keras
from tensorflow.keras import layers

# Keras Sequential — her şey otomatik
model = keras.Sequential([
    layers.Input(shape=(2,)),
    layers.Dense(4, activation='relu'),    # Gizli katman
    layers.Dense(1, activation='sigmoid'), # Çıkış
])

# Keras X'i (örnek_sayısı, özellik) bekler — bizim transposeumuz
X_keras = X_xor.T   # (4, 2)

# Tahmin (eğitilmemiş, rastgele ağırlıklarla)
tahminler = model.predict(X_keras, verbose=0)

print("Keras modeli — eğitilmemiş tahminler:")
for i in range(4):
    print(f"  ({X_keras[i,0]}, {X_keras[i,1]}) → {tahminler[i,0]:.3f}")

print("\n💡 Keras tüm matris işlemlerini, aktivasyonları arka planda yapıyor.")
print("   Üç satırlık model, bizim yazdığımız NumPy kodunun TIPATIP aynısı!")
````

## 🎮 Kendin Dene

Aşağıda kendi 2 girdili 3 katmanlı ağınızla oynayın. Sayıları değiştirip çıktıların nasıl etkilendiğini görün:

````python
# 👇 İstediğiniz değerleri girin
girdi_1 = 0.7
girdi_2 = -0.3

# Ağırlıkları sabit bıraktık (rastgele üretilmiş W1, W2 yukarıdan kullanılıyor)
x_dene = np.array([[girdi_1], [girdi_2]])
z1 = W1 @ x_dene + b1
a1 = relu(z1)
z2 = W2 @ a1 + b2
a2 = sigmoid(z2)

print(f"Girdi: ({girdi_1}, {girdi_2})")
print(f"Gizli katman çıktıları (a1): {a1.flatten().round(3)}")
print(f"Final tahmin: {a2[0,0]:.4f}")
````

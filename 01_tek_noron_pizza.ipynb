## 🤔 Önce Bir Senaryo

Bir pizzacı düşünün. **Pizza ne kadar büyükse, o kadar pahalıdır** — bu bir kural.

| Pizza Çapı (cm) | Fiyat (₺) |
|---|---|
| 20 | 100 |
| 25 | 130 |

**Soru:** Çapı 27 cm olan pizza kaç lira olur?

İnsan beyni bu soruya hemen cevap verebilir: "Ortalamaya bakıp tahmin ederim, 140 civarı olur." Peki **bilgisayara bunu nasıl öğretiriz?**

İşte cevap: **TEK BİR NÖRON** ile.

---

### 💡 Nöron Nedir?

Nöron, sadece şu işi yapan bir **mini makinedir**:

$$\text{tahmin} = (\text{ağırlık} \times \text{girdi}) + \text{bias}$$

Veya kısaca:  $$y = w \cdot x + b$$

- **w (ağırlık):** Girdiyi ne kadar "önemseyeceğini" söyler. Çapın fiyatı ne kadar etkilediği.
- **b (bias):** Sabit bir kayma. Sıfır çaplı pizza bile bir tabak parası tutar 🍽️
- **x (girdi):** Pizza çapı.
- **y (tahmin):** Fiyat tahmini.

Bizim işimiz: **w ve b'nin doğru değerlerini bulmak.** Bu işleme **"öğrenme"** diyoruz.

> 🎓 **Analoji:** Terazi düşünün. İki tarafa farklı ağırlıklar koyup denge buluyorsunuz. Nöron da ağırlık (w) ve bias (b) değerlerini denge bulana kadar ayarlar.


## 📦 Kurulum

İki kütüphane yeterli:

- **NumPy:** Sayılarla hızlı matematik işlemleri için
- **Matplotlib:** Grafik çizmek için

> 💬 Google Colab kullanıyorsanız bunlar **zaten yüklü**. Aşağıdaki hücreyi çalıştırmanız yeterli.

```python
import numpy as np
import matplotlib.pyplot as plt

# Grafiklerin daha güzel görünmesi için
plt.rcParams['figure.figsize'] = (9, 5)
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.alpha'] = 0.3

print("✅ Kütüphaneler hazır!")
```

## 📊 Veriyi Tanıyalım

Pizzacımızdan topladığımız 10 örnekli veri:

```python
# Pizza çapları (cm)
caplar = np.array([20, 22, 24, 26, 28, 30, 32, 34, 36, 38], dtype=float)

# Gerçek fiyatlar (₺)
fiyatlar = np.array([95, 105, 122, 138, 155, 170, 185, 200, 215, 230], dtype=float)

# Tabloyu yazdıralım
print("Çap (cm) | Fiyat (₺)")
print("-" * 22)
for c, f in zip(caplar, fiyatlar):
    print(f"   {c:>5.0f} |  {f:>5.0f}")
```

### 🎨 Veriyi Görselleştirelim

Sayılara bakmak yerine **noktalı bir grafik** çizelim — kalıbı görmek çok daha kolay olur.

```python
plt.scatter(caplar, fiyatlar, color='#1E88E5', s=80, edgecolor='white', linewidth=2, zorder=3)
plt.xlabel('Pizza Çapı (cm)', fontsize=12)
plt.ylabel('Fiyat (₺)', fontsize=12)
plt.title('🍕 Pizza Çapı vs Fiyatı', fontsize=14, fontweight='bold')
plt.show()

print("📈 Gözlem: Veriler neredeyse düz bir çizgide diziliyor!")
print("    Yani çap arttıkça fiyat da DOĞRUSAL şekilde artıyor.")
print("    İşte tek nöron tam bu tür kalıpları yakalar.")
```

## 🧠 Tek Nöron Modelimizi Kuralım

Hatırlayın: $y = w \cdot x + b$

**Başlangıçta** w ve b'yi rastgele seçiyoruz (sanki gözü kapalı tahmin yapıyormuş gibi).

```python
# Rastgele başlangıç değerleri
np.random.seed(42)  # Aynı sonuçları görmek için
w = np.random.randn() * 0.1   # Küçük rastgele bir sayı
b = np.random.randn() * 0.1

print(f"Başlangıç ağırlık (w) : {w:.4f}")
print(f"Başlangıç bias (b)    : {b:.4f}")
```

### Nöron'u Tanımlayalım

Şimdi tahmin yapan tek satırlık fonksiyon:

```python
def noron_tahmin(x, w, b):
    """Tek nöronun tahmini: y = w*x + b"""
    return w * x + b

# Bakalım eğitilmemiş nöron 30 cm pizza için ne tahmin ediyor?
ilk_tahmin = noron_tahmin(30, w, b)
print(f"Eğitilmemiş nöronun 30 cm pizza tahmini: {ilk_tahmin:.2f} ₺")
print(f"Gerçek fiyat ise: 170 ₺")
print()
print("😅 Tahmin çok kötü çünkü w ve b henüz rastgele!")
```

### 📉 Hata Nasıl Ölçülür?

Tahminimizin ne kadar kötü olduğunu **sayıyla ifade etmek** zorundayız ki düzeltebilelim.

En yaygın yöntem: **Ortalama Kare Hata (MSE — Mean Squared Error)**

$$\text{Hata} = \text{Ortalama}((\text{tahmin} - \text{gerçek})^2)$$

> 🎓 **Neden kare?** Çünkü tahmin gerçek değerin altında da olabilir, üstünde de. Eksi/artı işaretler birbirini yutmasın diye kareliyoruz.


```python
def hata_hesapla(tahminler, gercekler):
    """Ortalama kare hata"""
    return np.mean((tahminler - gercekler) ** 2)

# Tüm pizzalar için tahmin yap
tum_tahminler = noron_tahmin(caplar, w, b)
ilk_hata = hata_hesapla(tum_tahminler, fiyatlar)

print(f"Eğitimsiz nöronun toplam hatası: {ilk_hata:.2f}")
print("(Bu sayı küçüldükçe model iyileşir)")
```

## 🚀 Şimdi Asıl Olay: Öğrenme!

**Öğrenme şu demek:** w ve b değerlerini, hatayı azaltacak şekilde **küçük adımlarla** değiştirmek.

Bu işlemin matematiksel adı: **Gradyan İniş (Gradient Descent)**

> 🎓 **Analoji:** Gözleri kapalı tepeden inmek gibi. Her adımda ayağınızla "aşağı eğim hangi tarafta?" diye yokluyorsunuz, o yöne küçük bir adım atıyorsunuz.

Her döngüde (her **epoch**'ta) şunu yapacağız:

1. Mevcut w, b ile tahmin yap
2. Hatayı ölç
3. w ve b'yi hatayı azaltacak yönde **biraz** güncelle
4. Tekrar et

```python
# Eğitim parametreleri
ogrenme_orani = 0.0005   # Adımı ne kadar büyük atalım? (çok büyükse atlar, çok küçükse yavaş)
epoch_sayisi = 200       # Kaç tur dönelim?

# Her tur'daki hatayı saklayalım
hata_gecmisi = []

# w ve b'yi sıfırlayıp baştan başlayalım
w, b = 0.0, 0.0

for epoch in range(epoch_sayisi):
    # 1. Tahmin yap
    tahminler = noron_tahmin(caplar, w, b)

    # 2. Hatayı kaydet
    hata = hata_hesapla(tahminler, fiyatlar)
    hata_gecmisi.append(hata)

    # 3. Gradyanları hesapla (matematik bilmenize gerek yok, formül bu)
    dw = np.mean(2 * (tahminler - fiyatlar) * caplar)
    db = np.mean(2 * (tahminler - fiyatlar))

    # 4. w ve b'yi güncelle (azalan yönde küçük bir adım)
    w = w - ogrenme_orani * dw
    b = b - ogrenme_orani * db

    # Her 20 adımda bir bilgi yazdır
    if epoch % 20 == 0:
        print(f"Epoch {epoch:>3} | Hata: {hata:>8.2f} | w: {w:.3f} | b: {b:.3f}")

print()
print(f"🎉 Eğitim bitti!")
print(f"   Son w: {w:.3f}")
print(f"   Son b: {b:.3f}")
print(f"   Son hata: {hata_gecmisi[-1]:.2f}")
```

### 📈 Eğitim Sürecini Görelim

İki grafik çizeceğiz:
1. **Hata grafiği:** Her epoch'ta hata azalmalı
2. **Çizgi uydurma:** Eğitilmiş nöronun çizgisi noktaların ortasından geçmeli

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 1. Grafik: Hata zamanla nasıl değişti?
axes[0].plot(hata_gecmisi, color='#E53935', linewidth=2)
axes[0].set_xlabel('Epoch (Eğitim Turu)', fontsize=11)
axes[0].set_ylabel('Hata (MSE)', fontsize=11)
axes[0].set_title('📉 Hata Azalıyor — Nöron Öğreniyor!', fontsize=13, fontweight='bold')

# 2. Grafik: Veriler ve nöronun bulduğu çizgi
axes[1].scatter(caplar, fiyatlar, color='#1E88E5', s=80, edgecolor='white',
                linewidth=2, zorder=3, label='Gerçek Fiyatlar')
x_line = np.linspace(caplar.min() - 2, caplar.max() + 2, 100)
y_line = noron_tahmin(x_line, w, b)
axes[1].plot(x_line, y_line, color='#43A047', linewidth=3, label=f'Nöronun Çizgisi: y = {w:.2f}x + {b:.2f}')
axes[1].set_xlabel('Pizza Çapı (cm)', fontsize=11)
axes[1].set_ylabel('Fiyat (₺)', fontsize=11)
axes[1].set_title('🎯 Nöron Çizgiyi Bulmuş!', fontsize=13, fontweight='bold')
axes[1].legend(fontsize=11)

plt.tight_layout()
plt.show()
```

## 🔍 Test Edelim — Nöronumuz Ne Kadar İyi?

Şimdi eğitilmiş nöronumuza birkaç **yeni** çap soralım:

```python
test_caplari = [27, 33, 40, 50]

print("Çap (cm) | Nöronun Tahmini | Yorum")
print("-" * 50)
for cap in test_caplari:
    tahmin = noron_tahmin(cap, w, b)
    print(f"   {cap:>5} cm | {tahmin:>8.2f} ₺")

print()
print("💡 Dikkat: 50 cm bizim eğitim verimizde YOK (en büyük 38 cm idi),")
print("   ama nöron yine de mantıklı bir tahmin yapıyor.")
print("   Çünkü doğrusal kalıbı yakaladı!")
```

## 🎮 Kendin Dene

Aşağıdaki hücrede `cap` değişkenini değiştirip kendi pizzanızın fiyatını tahmin ettirin:

```python
# 👇 Buraya istediğiniz çapı yazın (cm cinsinden)
cap = 28

tahmin = noron_tahmin(cap, w, b)
print(f"🍕 {cap} cm pizza için tahmini fiyat: {tahmin:.2f} ₺")
```

## 📝 Özet — Bu Notebook'ta Ne Öğrendik?

| Terim | Türkçe Karşılığı | Pizza Örneğindeki Karşılığı |
|---|---|---|
| **Neuron** | Nöron | $y = w \cdot x + b$ formülünü hesaplayan birim |
| **Weight (w)** | Ağırlık | Çapın fiyatı ne kadar etkilediği |
| **Bias (b)** | Sapma / Sabit | Çap sıfır olsa bile fiyatın başlangıç değeri |

| **Epoch** | Tur / Devir | Tüm verinin kaç kez gözden geçirildiği |
